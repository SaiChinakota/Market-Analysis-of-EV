```
df1 = spark.read.format("csv").option("header", "true").load("dbfs:/FileStore/shared_uploads/schinako@gmu.edu/Electric_Vehicle_Population_Data_2_.csv")
```

```
# Show the first 10 rows of the DataFrame
df1.show(10)
```

```
#Get the total number of electric vehicles registered in Washington State
total_electric_vehicles = df1.count()
total_electric_vehicles
```

```
# Get the top 10 most popular electric vehicle models in Washington State
top_10_models = df1.groupBy("model").count().orderBy("count", ascending=False).limit(10)
top_10_models.show()
```
